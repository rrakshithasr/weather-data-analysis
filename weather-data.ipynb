```python
# ==============================
# WEATHER DATA ANALYSIS DASHBOARD
# ==============================

import pandas as pd
import matplotlib.pyplot as plt

# Create Weather Dataset
weather = {
    "Day": ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"],
    "Temperature": [30, 32, 29, 31, 33, 28, 30],
    "Humidity": [65, 60, 70, 68, 58, 75, 66],
    "Rainfall": [0, 2, 10, 0, 0, 15, 5],
    "WindSpeed": [12, 10, 15, 11, 9, 18, 13]
}

df = pd.DataFrame(weather)

# Save CSV
df.to_csv("weather_data.csv", index=False)

# Read CSV
df = pd.read_csv("weather_data.csv")

# ------------------------------
# Data Analysis
# ------------------------------
print("="*40)
print(" WEATHER DATA ANALYSIS")
print("="*40)

print("\nDataset:\n")
print(df)

print("\nAverage Temperature:", round(df["Temperature"].mean(),2), "°C")
print("Highest Temperature:", df["Temperature"].max(), "°C")
print("Lowest Temperature:", df["Temperature"].min(), "°C")
print("Average Humidity:", round(df["Humidity"].mean(),2), "%")
print("Total Rainfall:", df["Rainfall"].sum(), "mm")
print("Maximum Wind Speed:", df["WindSpeed"].max(), "km/h")

print("\nHighest Temperature Day:",
      df.loc[df["Temperature"].idxmax(),"Day"])

print("Rainiest Day:",
      df.loc[df["Rainfall"].idxmax(),"Day"])

# ------------------------------
# Dashboard
# ------------------------------
plt.figure(figsize=(16,10))

# 1 Line Chart
plt.subplot(2,3,1)
plt.plot(df["Day"], df["Temperature"], marker='o')
plt.title("Temperature")
plt.xlabel("Day")
plt.ylabel("°C")
plt.grid(True)

# 2 Bar Chart
plt.subplot(2,3,2)
plt.bar(df["Day"], df["Humidity"])
plt.title("Humidity")
plt.xlabel("Day")
plt.ylabel("%")
plt.xticks(rotation=30)

# 3 Scatter Plot
plt.subplot(2,3,3)
plt.scatter(df["Temperature"], df["Rainfall"])
plt.title("Temperature vs Rainfall")
plt.xlabel("Temperature")
plt.ylabel("Rainfall")

# 4 Pie Chart
plt.subplot(2,3,4)
plt.pie(df["Rainfall"], labels=df["Day"], autopct="%1.1f%%")
plt.title("Rainfall Distribution")

# 5 Histogram
plt.subplot(2,3,5)
plt.hist(df["WindSpeed"], bins=5)
plt.title("Wind Speed Distribution")
plt.xlabel("Wind Speed")
plt.ylabel("Frequency")

# 6 Combined Line Chart
plt.subplot(2,3,6)
plt.plot(df["Day"], df["Temperature"], marker='o', label="Temperature")
plt.plot(df["Day"], df["Humidity"], marker='s', label="Humidity")
plt.title("Temperature & Humidity")
plt.xlabel("Day")
plt.legend()
plt.grid(True)

plt.tight_layout()

# Save Dashboard
plt.savefig("weather_dashboard.png")

plt.show()

print("\nDashboard saved as 'weather_dashboard.png'")
print("CSV file saved as 'weather_data.csv'")
```
